````python
import numpy as np

expected_rewards = np.array([1, 6, 3])

discount_factor = 0.9

discounts = np.array([discount_factor ** i for i in range(len(expected_rewards))])

print(discounts)
````

````python
discounted_return = np.sum(expected_rewards*discounts)
print(discounted_return)
````

````python
import gymnasium as gym

env = gym.make('CartPole', render_mode='rgb_array') #render_mode helps us visualise via matplotlib

state, info = env.reset(seed=42) #reset helps init env + auxillary info
print(state) #the observation is an array representing the environment's state, including the position and velocity of both the cart and the pole
````

````python
#for visualisation of the state

import matplotlib.pyplot as plt

def render():
    state_image = env.render()

    plt.imshow(state_image)

    plt.show()

render()
````

````python
# actions
# 0 = moving left
# 1 = moving right

action = 1

state, reward, terminated, truncated, info = env.step(action) #terminated = whether agent has reached goal or terminal signal; truncated = signal whether condition like time limit is met
#info = auxillary info useful for debugging

print(state)
print(reward) # = 1 means agent has not yet reached terminal state
print(terminated)
````

````python
#lets, we want to push the cart to the right until the terminal state is met

while not terminated:
    action = 1
    state, reward, terminated, _, _ = env.step(action)
    render()
````
